```python
import matplotlib.pyplot as plt
import numpy as np
```

# Hist

```python
np.random.seed(42)
data = [1,5,10,50] # 1000 داده با توزیع نرمال

plt.hist(data, bins=10, color='blue', edgecolor='black',rwidth=0.8)
plt.xlabel('Data')
plt.ylabel('Tedad')
plt.show()
```

# Bar

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(labels, values, color=['red', 'blue'],width=1.2)
plt.xlabel('Categories')
plt.ylabel('Value')
plt.show()
```

# Scatter

```python
x = [0,1,2,3,4]
y = [5,10,15,20,25]

plt.scatter(x, y, alpha=1) # alpha : opacity
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Pie

```python
sizes = [300, 20, 250, 750]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors,startangle=90)
plt.show()
```

# Plot Limit

```python
a = [1,2,3,4,5]
b = [10,20,30,40,50]

plt.ylim(top=100)
plt.xlim(right=10)
plt.plot(a,b)
```

```python
plt.ylim(-100,100)
plt.xlim(-10,10)
plt.plot(a,b)
```
